# Order 30779295

This order contains a Trade Fill Cancel sequence where the Fill occurs on the SELL side at a lower price than the Cancel.

- The Cancel message implies that there is a resting order in the book.
- This is presumably a limit order.
- How can it be that the limit order on the SELL side is filled at a lower price than the limit price?
